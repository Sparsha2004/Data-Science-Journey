# Task 5 - Simple Calculator Menu
This program allows the user to perform one operation at a time.

```python
#simple Calculator Menu
num1 = int(input("Enter a number: "))
num2 = int(input("Enter another number: "))
op = input("Enter the operation you want to perform (+,-,*,/):")
if (op == '+'):
    print(num1+num2)
elif (op == '-'):
    print(num1-num2)
elif (op == '*'):
    print(num1*num2)
elif (op == '/'):
    print(num1/num2)
else:
    print("Invalid operator.")
    
```
