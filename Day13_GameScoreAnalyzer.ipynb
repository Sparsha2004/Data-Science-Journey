```python
#Creating column in a dataset(logic)
players = ["Ayan", "Bikash", "Chirag", "Deepa", "Esha", "Farhan"]
scores  = [120,    150,     200,      120,     200,    150]
game = {}
for i in range(0, len(players)):
    game[players[i]] = scores[i]
for key in game:
    print(f"{key} : {game[key]}")
```

```python
#Score Distribution (Frequency Analysis- Score Distribution, Value Count, Categorical Frequency)
ScoreFreq = {}
for key in game:
    if game[key] in ScoreFreq:
        ScoreFreq[game[key]] += 1
    else:
        ScoreFreq[game[key]] = 1
print(ScoreFreq)
```

```python
#Finding Mode Scores (handling uniform distributions, computing mode with ties)
MaxFreq = 0
MaxFreqList = []
for key in ScoreFreq:
    if ScoreFreq[key] > MaxFreq:
        MaxFreq = ScoreFreq[key]
for key in ScoreFreq:
    if ScoreFreq[key] == MaxFreq:
        MaxFreqList.append(key)
print(f"Most common score(s): {MaxFreqList}")
print(f"Frequency: {MaxFreq}")
```

```python
#Player Grouping (groupby, categorical aggregation, one-to-many mapping)
revGame = {}
for key in game:
        if game[key] in revGame:
            revGame[game[key]].append(key)
        else:
            revGame[game[key]] = [key]
print(f"{revGame}")
```

```python
#Flattening grouped data (Filtering, Reporting, Feature Selection)
add = 0
HighScorers = []
for i in scores:
    add += i
average = add/len(scores)
for key in revGame:
    if key > average:
        for player in revGame[key]:
            HighScorers.append(player)
print(f"The average score is: {round(average,2)} and the players that scored above average are: {HighScorers}")
```
