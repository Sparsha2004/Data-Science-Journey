#List Basics

```python
intlist = [12,34,56,67,32,76,34,90,7,34]
#printing all list elements
for i in intlist:
    print(i, end = " ")
```

```python
#printing only even elements
for i in intlist: 
    if (i % 2 == 0):
        print(i, end = " ")
```

```python
#printing elements with even indices
for i in range(0, len(intlist)):
    if (i % 2 == 0):
        print(intlist[i], end = " ")
```

```python
#counting number of elements in list without using len
count = 0
for i in intlist:
    if (i != '/0'):
        count += 1
print(f"There are {count} elements in the list.")
```

```python
#sum all elements using a loop
add = 0
for i in intlist:
    add += i
print(f"The sum of all elements in the list is: {add}")
```

```python
#Reversing a list without using reverse
l = len(intlist)
revlist = [None]*l
for i in range(0, l):
    revlist[i] = intlist[l-1-i]
print(revlist)
```

```python
#Copying list element by element
newlist = []
for i in intlist:
    newlist.append(i)
print(newlist)
```

```python
#Remove all occurences of a given number
for i in intlist:
    if (i == 34):
        newlist.remove(i)
print(newlist)
```

```python
#count how many times each element appears 
fruitlist = ['apple', 'orange', 'banana', 'banana', 'grapes', 'banana', 'apple', 'pineapple']
newfruitlist = []
for i in range(0, len(fruitlist)):
    count = 0
    if ((fruitlist[i] in newfruitlist) == False):
        newfruitlist.append(fruitlist[i])
        for j in fruitlist: 
            if (j == fruitlist[i]): 
                count += 1
        print(f"There are {count} {fruitlist[i]}s.")
```
