```python
#Nested Dictionary
orders = ["Veg", "Chicken", "Buff", "Veg", "Chicken"]
price  = [120,   150,       180,    120,   150]
qty    = [2,     3,         1,      1,     2]
menu = {}
for i in range(0, len(orders)):
    menu[i] = {"type" : orders[i], "price": price[i], "qty": qty[i]}
for key in menu:
    print(f"{key} : {menu[key]}")
```

```python
#Feature Engineering
revenue = []
for i in range(0, len(menu)):
    revenue.append(price[i] * qty[i])
    menu[i] = {"type" : orders[i], "price": price[i], "qty": qty[i], "revenue" : revenue[i]}
for key in menu:
    print(f"{key} : {menu[key]}")
```

```python
#Grouping and Aggregation (groupby().sum() in pandas)
MomoRev = {}
for i in menu:
    if orders[i] in MomoRev:
        MomoRev[orders[i]] += revenue[i]
    else:
        MomoRev[orders[i]] = revenue[i]
print(MomoRev)
```

```python
#Threshold Based Filtering (df[df['revenue'] > 400)
filteredDict = {}
for key in MomoRev:
    if MomoRev[key] > 400:
        filteredDict[key] = MomoRev[key]
print(f"{filteredDict}")
```

```python
#Combined Insights
MomoQty = {}
for i in menu:
    if orders[i] in MomoQty:
        MomoQty[orders[i]] += qty[i]
    else:
        MomoQty[orders[i]] = qty[i]
MomoPrice = {}
for i in menu:
        MomoPrice[orders[i]] = price[i]
Combined = {}
for key in MomoPrice:
    Combined[key] = {"total_qty": MomoQty[key], "total_revenue": MomoRev[key], "avg_price": MomoPrice[key]}
for key in Combined:
    print(f"{key} : {Combined[key]}")
```
