```python
#Using dictionary to solve frequency problem
marks = [45, 78, 90, 45, 50, 90, 78]
count = {}
for i in marks:
    if i in count:
        count[i] += 1
    else:
        count[i] = 1
for i in count:
    print(f"{i} appears {count[i]} times.")
```

```python
#counting number of letters in the word
word = "banana"
wordCount = {}
for i in word:
    if i in wordCount:
        wordCount[i] += 1
    else:
        wordCount[i] = 1
for i in wordCount:
    print(f"{i} appears {wordCount[i]} times.")
```

```python
#Mixed Challenge
numbers = [1, 2, 2, 3, 4, 4, 4, 5]
count = {}
for i in numbers:
    if i in count:
        count[i] += 1
    else:
        count[i] = 1
print(count)
maxIndex = 0
for i in count:
    if (count[i+1]) > count[i]:
        maxCount = (count[i+1])
        maxIndex = (i+1)
print(f"{maxIndex} appears {maxCount} times.")
    
```
