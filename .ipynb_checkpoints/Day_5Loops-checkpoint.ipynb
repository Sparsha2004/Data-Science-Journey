#List of Prime Numbers

```python
num = int(input("Enter a number: "))
print("List of prime numbers: ")
for j in range(1, num+1):
    count = 0
    for i in range(1,num+1):
        if (j % i == 0):
            count += 1
    if (count <= 2):
        print(j, end = " ")
```

#GCD finder

```python
num1 = int(input("Enter a number: "))
num2 = int(input("Enter another number: "))
while (num1 % num2 != 0):
    temp = num1 % num2
    num1 = num2
    num2 = temp
print("The value of GCD is: ", temp)
```

#Armstrong Number Checker

```python
num = int(input("Enter a number: "))
checknum = num
count = 0
while (checknum != 0):
    rem = checknum % 10
    checknum = checknum//10
    count += 1
checknum = num
newnum = 0
for i in range(1,count+1):
    rem = checknum % 10
    checknum = checknum//10
    print(f"{rem}^{count} = ", end = " ")
    newnum += (rem**count)
    print(f"{rem**count} ", end = " ")
print("=",newnum)
if (newnum == num):
    print("Armstrong number.")
else:
    print("Not armstrong number.")
```

#Counting number of digits

```python
num = int(input("Enter a number: "))
digits = str(num)
count = len(digits)
for j in range (0, 10):
    countd = 0
    for i in digits:
        rem = int(i)
        if (rem == j):
            countd += 1
    print(f"Number of digit {j} in the number you entered is: ", countd)       
```
