```python
mylist = [12, 45, 645, 75, 2, 6, 775, 234, 78, 8]

if len(mylist) < 2:
    print("List must have at least two elements.")
else:
    maxelement = mylist[0]
    secondmax = float("-inf")

    for x in mylist:
        if x > maxelement:
            secondmax = maxelement
            maxelement = x
        elif x < maxelement and x > secondmax:
            secondmax = x

    print("The second largest element is:", secondmax)
```

```python
newlist = [1, 3, 6, 3, 7, 3, 2, 7, 8, 8, 4]
newerlist = []
for i in newlist:
        if ((i not in newerlist):
            newerlist.append(i)
for j in newerlist:
    count = 0
    for k in newlist:
        if(j == k):
            count += 1
    print(f"{j} appears {count} number of times in the list.")
```
