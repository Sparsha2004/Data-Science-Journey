```python
mylist = [90, 25, 543, 23, 64, 26, 234, 632, 677, 168]
sqrlist = []
for i in mylist:
    i2 = i**2
    sqrlist.append(i2)
print("The list of squared numbers is: ",sqrlist)
```

```python
list2 = [4, 5, 6, 7, 3, 7, 3, 9, 2, 6, 2]
newlist = []
for i in list2:
    if ((i in newlist) == False):
        newlist.append(i)
print(newlist)
```
