#Print numbers from 1 to n

```python
n = int(input("Enter a number: "))
for val in range(1,n):
    print(val)
```

#Print sum of numbers from 1 to 10

```python
n = int(input("Enter a number: "))
i = 0
newnum = 0
while i <= n:
    newnum = i+newnum
    i+=1
print(newnum)
```

#Print even numbers between 1 and n

```python
n = int(input("Enter a number: "))
for val in range(1,n+1):
    if (val % 2 == 0):
        print (val)
        
```

#Multiplication table of a given number

```python
n = int(input("Enter a number: "))
for val in range(1,11):
    print(f"{n}X{val}={n*val}")
```

#Print numbers from 1 to n except multiples of 3

```python
n = int(input("Enter a number: "))
for val in range(1,n+1):
    if (val % 3 != 0):
        print (val)
        
```

#Print numbers from n to 1

```python
n = int(input("Enter a number: "))
for val in range(1, n+1):
    print(n)
    n-=1
    
```

#Count numbers divisible by 5

```python
n = int(input("Enter a number: "))
count = 0
for val in range(1, n+1):
    if(val%5==0):
        count+=1
print(count)
```

#Printing various patterns

```python
stars = "*"
for i in range(1,6):
    print(stars)
    stars = stars + "*"
```

```python
num = "12345"
for i in range(1,6):
    print(num[0:i])
    
```

```python
stars = "*****"
for i in range(1,6):
    j = 6-i
    print(stars[0:j])
```
