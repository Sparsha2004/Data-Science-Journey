#Advanced Logic+Lists

```python
mylist = [7, 18, 22, 31, 42, 67, 69, 100, 405, 420]
greatest = max(mylist)
secondGreat = 0
for index in range(0, len(mylist)):
    for i in mylist:
            if (i < greatest and i > mylist[index]):
                secondGreat = i
print("The second greatest number in the list is: ", secondGreat)
```

```python
sortedlist = sorted(mylist)
print(sortedlist)
if (sortedlist == mylist):
    print("The list is sorted.")
else:
    print("The list is not sorted.")
```

```python
evenlist = []
oddlist = []
for i in range(0, len(mylist)):
    if (mylist[i] % 2 == 0):
        evenlist.append(mylist[i])
    else:
        oddlist.append(mylist[i])
print("The list of even numbers from given list is:" ,evenlist)
print("The list of odd numbers from given list is:" ,oddlist)
```

```python
animalList = ['kangaroo', 'ostrich', 'apes','giraffe','hippo', 'kiwi']
fruitList = ['apple', 'kiwi', 'orange', 'guava', 'banana', 'grapes', 'watermelon']
mergeList = []
mergeList.extend(animalList)
mergeList.extend(fruitList)
print(mergeList)
```

```python
commonList = []
for i in range(0, len(animalList)):
    for j in range(0, len(fruitList)):
        if (animalList[i] == fruitList[j]):
            commonList.append(animalList[i])
print(commonList)
```
