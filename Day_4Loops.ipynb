#Counting the total number of numbers entered

```python
num = int(input("Enter a number: "))
rem = num%10
count = 0
while  rem != 0:
    num = num//10
    rem = num%10
    count += 1
print("Total number of digits = ", count)
```

#Reversing a number

```python
num = int(input("Enter a number: "))
newnum = 0
while num != 0:
    rem = num%10
    num = num//10
    newnum = newnum*10+rem
print(newnum)
```

#Fibonacci Series

```python
num = int(input("Enter the number you want to generate Fibonacci series of: "))
i = 0
print(i, end = " ")
j = 1
print(j, end = " ")
for k in range(3, num):
    temp = i + j
    i = j
    j = temp
    print(j, end = " ")


```

#Check if a number is palindrome

```python
num = int(input("Enter a number: "))
gum = num
newnum = 0
while gum != 0:
    rem = gum%10
    gum = gum//10
    newnum = newnum*10+rem
if(num == newnum):
    print("Palindrome")
else:
    print("Not palindrome")
```

#sum of digits of a number

```python
num = int(input("Enter a number: "))
add = 0
while num != 0:
    rem = num%10
    num = num//10
    add += rem
print("The sum of digits of the number is: ", add)
```
