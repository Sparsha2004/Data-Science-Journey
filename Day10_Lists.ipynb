```python
NumList = [49, 80, 45, 75, 28, 8, 24, 75, 16]
add = 0
for i in NumList:
    add += i
avg = add/len(NumList)
print(f"Numbers greater than average in the list {NumList} are: ")
GreaterList = []
for i in NumList:
    if (i > avg):
        GreaterList.append(i)
print(GreaterList)
print(NumList)
```

```python
SortCheck = [3, 7, 9, 15, 21]
isSorted = True
for i in range(0, len(SortCheck)-1):
    if SortCheck[i] > SortCheck[i+1]:
        isSorted = False
        break
print(SortCheck)
if (isSorted == True):
    print("Sorted.")
else:
    print("Not sorted.")
```
