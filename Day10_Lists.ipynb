```python
NumList = [49, 80, 45, 75, 28, 8, 24, 75, 16]
add = 0
for i in NumList:
    add += i
avg = add/len(NumList)
print(f"Numbers greater than average in the list {NumList} are: ")
GreaterList = []
for i in NumList:
    if (i > avg):
        GreaterList.append(i)
print(GreaterList)
print(NumList)
```

```python
SortCheck = [3, 7, 9, 15, 21]
isSorted = True
for i in range(0, len(SortCheck)-1):
    if SortCheck[i] > SortCheck[i+1]:
        isSorted = False
        break
print(SortCheck)
if (isSorted == True):
    print("Sorted.")
else:
    print("Not sorted.")
```

```python
marks = [45, 78, 90, 45, 50, 90, 78]
#Finding the average and printing above average marks
add = 0
for i in range(0, len(marks)):
    add += marks[i]
average = add/len(marks)
print(f"The average marks is: {average}")
greater_marks = []
print("The marks above average are:")
for k in marks:
    if k > average:
        greater_marks.append(k)
print(greater_marks)
#Finding unique marks and counting frequency
uniq_marks = []
for i in marks:
    if i not in uniq_marks:
        uniq_marks.append(i)
print(f"The unique marks from given marks are: {uniq_marks}.")
for i in uniq_marks:
    count = 0
    for j in marks:
        if i == j:
            count += 1
    print(f"{i} appears {count} number of time(s).")
#Finding if the list is sorted
isSorted = True
for i in range(0, len(marks)-1):
    if marks[i] > marks[i+1]:
        isSorted = False
        break
if isSorted == True:
    print("The list is sorted in ascending order.")
else:
    print("The list is not sorted in ascending order.")
```
