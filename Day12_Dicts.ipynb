```python
#Frequency Table (logic for pandas.value_counts())
data = [1, 2, 1, 3, 2, 4, 1]
freq = {}
for i in data:
    if i in freq:
        freq[i] += 1
    else:
        freq[i] = 1
for key in freq:
    print (f"{key} -> {freq[key]}")
```

```python
#Maximum Frequency List (logic for max(), value_counts(), idxmax() in pandas)
FreqDict = []
maxFreq = 0
for key in freq:
    if freq[key] > maxFreq:
        maxFreq = freq[key]
for key in freq:
    if freq[key] == maxFreq:
        maxFreqDict.append(key)
print(f"The values that have the maximum frequency are: {maxFreqDict} with frequency {maxFreq}.")
```

```python
#Dictionary from given values (df.grouby("marks") logic)
students = ["Alice", "Bob", "Charlie", "Diana", "Eve", "Frank"]
marks    = [78,     82,    91,        78,      91,    82]
report = {}
for i in range(0, len(students)):
    if marks[i] in report:
        report[marks[i]].append(students[i])
    else:
        report[marks[i]] = [students[i]]
print(report)
```

```python
#Reverse Mapping (dictionary inversion)
revReport = {}
for key in report:
    for val in report[key]:
        revReport[val] = key 
print(revReport)
```

```python
#Filtered Grouping (df[df["marks"] >= 80])
highResult = {}
for key in report:
    if key >= 80:
        highResult[key] = report[key]
print(highResult)
```
