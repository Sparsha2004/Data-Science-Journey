```python
#Using dictionary to solve frequency problem
marks = [45, 78, 90, 45, 50, 90, 78]
count = {}
for i in marks:
    if i in count:
        count[i] += 1
    else:
        count[i] = 1
for i in count:
    print(f"{i} appears {count[i]} times.")
```

```python
#counting number of letters in the word
word = "banana"
wordCount = {}
for i in word:
    if i in wordCount:
        wordCount[i] += 1
    else:
        wordCount[i] = 1
for i in wordCount:
    print(f"{i} appears {wordCount[i]} times.")
```

```python
#Mixed Challenge
numbers = [1, 2, 2, 3, 4, 4, 4, 5]
count = {}
for i in numbers:
    if i in count:
        count[i] += 1
    else:
        count[i] = 1
print(count)
maxCount = 0
maxNumber = 0
for key in count:
    if (count[key]) > maxCount:
        maxCount = count[key]
        maxNumber = key
print(f"{maxNumber} appears {maxCount} times.")
    
```

```python
#Student Exam Analysis Challenge
students = ["Alice", "Bob", "Charlie", "Diana", "Eve", "Frank"]
marks = [78, 82, 91, 78, 91, 82]
result = {}
#creating dictionary
for i in range(0, len(students)):
    result[students[i]] = marks[i]
print(result)
#counting frequency
markCount = {}
for key in result:
    if result[key] in markCount:
        markCount[result[key]] += 1
    else:
        markCount[result[key]] = 1
print(markCount)
#highest frequency
maxCount = 0
maxIndex = 0
for key in markCount:
    if markCount[key] > maxCount:
        maxCount = markCount[key]
        maxIndex = key
#Fixes-for-ties
mostFrequentMarks = []
for key in markCount:
    if markCount[key] == maxCount:
        mostFrequentMarks.append(key)
for m in mostFrequentMarks:
    print(f"Mark {m} is obtained by:")
    studentsWithMark = []
    for s in result:
        if result[s] == m:
            studentsWithMark.append(s)
    print(studentsWithMark)
#above average calculation
add = 0
highScorers = []
for i in marks:
    add += i
average = add/len(marks)
print(f"The students that scored above average({average}): ")
for key in result:
    if result[key] > average:
        highScorers.append(key)
print(highScorers)
```
