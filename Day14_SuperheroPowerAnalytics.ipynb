```python
#Building Nested Data (foundation)
heroes = ["IronMan", "Batman", "Thor", "Hulk", "Flash", "Superman"]
power  = [85,        80,       95,     98,     90,      99]
speed  = [70,        65,       60,     55,     98,      95]
superpower = {}
for i in range(0, len(heroes)):
    superpower[heroes[i]] = {"power" : power[i], "speed" : speed[i]}
for hero in superpower:
    print(f"{hero} : {superpower[hero]}")
```

```python
#Combined Power Score (ML-style thinking)
HeroScore = {}
combined_score = []
#Previous Attempt:
# for i in range(0, len(superpower)):
#     combined_score.append(power[i] + speed [i])
#     HeroScore[heroes[i]] = combined_score[i]
#Correct Attempt:
for hero in superpower:
    HeroScore[hero] = superpower[hero]["power"] + superpower[hero]["speed"]
for hero in HeroScore:
    print(f"{hero} : {HeroScore[hero]}")
```

```python
#Category Classification
HeroTier = {}
for hero in HeroScore:
    if HeroScore[hero] >= 180:
        HeroTier[hero] = "S-Tier"
    elif HeroScore[hero]<180 and HeroScore[hero]>=150:
        HeroTier[hero] = "A-Tier"
    else:
        HeroTier[hero] = "B-Tier"
for hero in HeroTier:
    print(f"{hero} : {HeroTier[hero]}")
```

```python
#Insight Extraction
fastHero = " "
fastSpeed = 0
for hero in HeroTier:
    if HeroTier[hero] == "S-Tier":
        if superpower[hero]["speed"] > fastSpeed:
            fastHero = hero
            fastSpeed = superpower[hero]["speed"]
print(f"Fastest S-Tier hero: {fastHero} with speed {fastSpeed}")
```

```python
#Weighted Ranking
w1 = 0.6
w2 = 0.4
heroRank = {}
for hero in superpower:
    heroRank[hero] = w1 * superpower[hero]["speed"] + w2 * superpower[hero]["power"]
for hero in heroRank:
    print(f"{hero} : {heroRank[hero]}")
```
